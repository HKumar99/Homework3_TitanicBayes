```python
import pandas as pd
locate = "titanic3.xls"
df = pd.read_excel(locate)
df.head()
```

```python
def gender_to_binary(x):
    if x == 'female':
        return 1
    else:
        return 0  
```

```python
df['sex'] = df['sex'].apply(gender_to_binary)
df.head()
```

```python
def relative_add(list1, list2):
    return list1 + list2
```

```python
df['relnum'] = relative_add(df['sibsp'], df['parch'])
df.head()
```

```python
df.drop_duplicates()
df.head()
```

```python
survived_df = df.drop(df[df['survived'] == 0].index)
survived_df.head()
```

```python
survived_df.head(10)
```

```python
def total_survived_woman(list1):
    total = 0;
    for w in list1:
        if (w == 1):
            total += 1;
    return total;

total_woman_survived = total_survived_woman(survived_df['sex'])
total_woman_survived
```

```python
total_survived = survived_df.shape[0]
total_survived
```

```python
prob_gender = total_woman_survived / total_survived
```

```python
def total_children_survived(list1):
    total = 0;
    for c in list1:
        if (c < 18):
            total += 1;
    return total;

total_survived_children = total_children_survived(survived_df['age'])
total_survived_children
```

```python
prob_age = total_survived_children / total_survived
prob_age
```

```python
first_total = 0;
second_total = 0;
third_total = 0;
def first_class_survived(list1):
    global first_total
    global second_total
    global third_total
    
    for f in list1:
        if (f == 1):
            first_total += 1;
        elif (f == 2):
            second_total += 1;
        elif (f == 3):
            third_total += 1;

blah = first_class_survived(survived_df['pclass'])

prob_first = first_total / total_survived;
prob_second = second_total / total_survived;
prob_third = third_total / total_survived;

print(prob_first)
print(prob_second)
print(prob_third)
```

```python
def no_of_woman(list1):
    num = 0;
    for w in list1:
        if (w == 1):
            num += 1;
    return num;

def no_of_child(list1):
    num = 0;
    for c in list1:
        if (c < 18):
            num += 1;
    return num;

def no_of_classone(list1):
    num = 0;
    for c in list1:
        if (c == 1):
            num += 1;
    return num;

def no_of_classtwo(list1):
    num = 0;
    for c in list1:
        if (c == 2):
            num += 1;
    return num;

def no_of_classthird(list1):
    num = 0;
    for c in list1:
        if (c == 3):
            num += 1;
    return num;

total_woman = no_of_woman(df['sex'])
total_child = no_of_child(df['age'])
total_one = no_of_classone(df['pclass'])
total_two = no_of_classtwo(df['pclass'])
total_third = no_of_classthird(df['pclass'])

total = df.shape[0]

print(total_woman)
print(total_child)
print(total_one)
print(total_two)
print(total_third)
print(total)
```

```python
prob_woman = total_woman / total
print(prob_woman)

prob_ch = total_child / total
print(prob_ch)

prob_o = total_one / total
print(prob_o)

prob_tw = total_two / total
print(prob_tw)

prob_th = total_third / total
print(prob_th)
```
